```python
data= [(1,2),(3,1),(0,4)]
sorted_by_first = sorted(data)  # Default sorts by the first element
print(sorted_by_first)
sorted_by_second = sorted(data, key=lambda x: x[1])
print(sorted_by_second)
```

```python
data= [(1,2),(3,1),(0,4)]
sorted_by_first = sorted(data)  # Default sorts by the first element
print(sorted_by_first)
sorted_by_second = sorted(data, key=lambda x: x[1],reverse=True)
print(sorted_by_second)
```

```python

data2 = [(1, 3), (1, 2), (2, 1)]
sorted_by_first_second = sorted(data2, key=lambda x: (x[0], x[1]))
print(sorted_by_first_second)
```

```python
data = { 'date': 1, 'apple': 5, 'banana': 2, 'cherry': 7}
sorted_by_key = dict(sorted(data.items()))
print(sorted_by_key)
```

```python
data = {'ba': 5, 'banana': 5, 'cherry': 7, 'date': 1}

s=sorted(data.items(),key=lambda x:(x[0],x[1]))
print(dict(s))
```

```python
data = {'ba': 5, 'banana': 5, 'cherry': 7, 'date': 1}

s=sorted(data.items(),key=lambda x:(x[1],x[0]))
print(dict(s))
```

```python
# Sort by keys in reverse order
sorted_by_key_desc = dict(sorted(data.items(), reverse=True))
print(sorted_by_key_desc)


# Sort by values (ascending)
sorted_by_value = dict(sorted(data.items(), key=lambda item: item[1]))
print(sorted_by_value)

# Sort by values (descending)
sorted_by_value_desc = dict(sorted(data.items(), key=lambda item: item[1], reverse=True))
print(sorted_by_value_desc)
```

```python
# Zip with Unequal Length Lists

list1 = [1, 2, 3, 4]
list2 = ['a', 'b']

zipped = zip(list1, list2)
print(list(zipped))
```

```python
# Zip More than Two Lists
list1 = [1, 2, 3]
list2 = ['a', 'b', 'c']
list3 = ['x', 'y', 'z']

zipped = zip(list1, list2, list3)
print(list(zipped))
```

```python
# 1. map

cel=[24,26,28,30]
fah=map(lambda x: x*1.6+32 , cel)
print(list(fah))

# 2. list comprehension

cel=[24,26,28,30]
fah=[x*1.6+32 for x in cel]
print(cel)
print(fah)
```

```python
class NegativeNumberError(Exception):     # Exception is predefined class. It is base class 

    pass                                  # pass does nothing

try:

    age=int(input("Enter value for age ")) 

    if age < 0:

        raise NegativeNumberError("Age with negative value is not allowed")

    else:

        print("Age is   " , age )

except NegativeNumberError as e:

    print("Custom Exception:", e)
 
 
```

```python
class NegativeNumberError(Exception):     # Exception is predefined class. It is base class 
    pass                                  # pass does nothing
try:
    age=int(input("Enter value for age ")) 
    if age < 0:
        raise NegativeNumberError("Age with negative value is not allowed")
    if age >=0 and age <18:
       raise NegativeNumberError("Age  must be above 18  ")
    if age>=18:
       print("My age is   ", age)
except NegativeNumberError as e:
    print("Custom Exception:", e)
```
